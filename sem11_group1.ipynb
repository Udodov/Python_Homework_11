## Задача 1

Дана функция f(x)=5x^2+10x-30

1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0  по y
7. Определить промежутки, на котором f < 0
<hr>

```python
import numpy as np
import matplotlib.pyplot as plt


x_range = np.arange(-10, 10.01, 0.01)
func = 5 * x_range**2 + 10 * x_range - 30
a, b, c = 5, 10, -30


def sqr_roots(a, b, c):
    discr = b**2 - 4 * a * c
    if discr > 0:
        x1 = (-b + np.sqrt(discr)) / (2 * a)
        x2 = (-b - np.sqrt(discr)) / (2 * a)
        return (x1, x2)
    elif discr == 0:
        x = -b / (2 * a)
        return x
    else:
        return


print(f"x1 = {round(sqr_roots(a,b,c)[0],2)}, x2 = {round(sqr_roots(a,b,c)[1],2)}")

min_func = min(func)

x = sqr_roots(a, b, c - min_func)

print(f"Вершина функции = {x}")


def change_func(range):
    func = 5 * range**2 + 10 * range - 30
    plt.plot(range, func)
    plt.show()


x_range1 = np.arange(-10, x, 0.01)
print("Убывание Функции")
change_func(x_range1)

x_range2 = np.arange(x, 10, 0.01)
print("Возрастание Функции")
change_func(x_range2)
```
