```python
import math


print("Введите коэффициенты для уравнения")
print("ax^2+bx+c=0:")
a = float(input("a= "))
b = float(input("b= "))
c = float(input("c= "))

discr = b**2 - 4 * a * c
print("Дискриминант D= %.2f" % discr)

if discr > 0:
    x1 = (-b + math.sqrt(discr)) / (2 * a)
    x2 = (-b - math.sqrt(discr)) / (2 * a)
    print("x1 = %.2f \nx2 = %.2f" % (x1, x2))
elif discr == 0:
    x = -b / (2 * a)
    print("x = %.2f" % x)
else:
    print("Коней нет")
```

```python
from sympy import *
import numpy as np

x = Symbol("x")
y = 5 * x**2 + 10 * x - 30

yprime = y.diff(x)
print(yprime)
```

```python
print(str(yprime).split("+"))
```

```python
x1 = solve(yprime, x)
print("Вершина: ", x1[0])
```

```python
up = yprime.subs(x, x1[0] + 1)
down = yprime.subs(x, x1[0] - 1)
print(up)
print(down)
```
