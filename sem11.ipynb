## Задача 1

Дана функция f(x)=5x^2+10x-30

1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0  по y
7. Определить промежутки, на котором f < 0
<hr>

Импортируем необходимые библиотеки

```python
import numpy as np
import matplotlib.pyplot as plt
```

Определяем функцию

```python
def f(x):
    return 5 * x**2 + 10 * x - 30
```

### 1. Определение корней
Функция `np.roots()` из библиотеки **NumPy** для нахождения корней квадратного уравнения, заданного коэффициентами `[5, 10, -30]`.</br> Корни сохраняются в переменной `roots`.

```python
roots = np.roots([5, 10, -30])
print("Корни функции:", roots)
```

### 2. Интервалы, на которых функция возрастает
Определяем интервалы, на которых функция возрастает. Если имеется два корня, то функция возрастает на интервалах между корнями и до первого корня и после второго корня. Если имеется один корень, то функция возрастает либо до корня (если корень отрицательный), либо после корня (если корень положительный). Если корней нет, то функция возрастает на всей числовой прямой. Интервалы сохраняются в переменной `increasing_intervals`.

```python
increasing_intervals = []
if len(roots) == 2:
    increasing_intervals.append((-np.inf, roots[0]))
    increasing_intervals.append((roots[1], np.inf))
elif len(roots) == 1:
    if roots[0] < 0:
        increasing_intervals.append((roots[0], np.inf))
    else:
        increasing_intervals.append((-np.inf, roots[0]))
else:
    increasing_intervals.append((-np.inf, np.inf))
print("Интервалы возрастания:", increasing_intervals)
```

### 3. Интервалы, на которых функция убывает
Если имеется два корня, то функция убывает на интервале между корнями. Если имеется один корень, то функция убывает либо до корня (если корень положительный), либо после корня (если корень отрицательный). Если корней нет, то функция убывает на всей числовой прямой. Интервалы сохраняются в переменной `decreasing_intervals`.

```python
decreasing_intervals = []
if len(roots) == 2:
    decreasing_intervals.append((roots[0], roots[1]))
elif len(roots) == 1:
    if roots[0] < 0:
        decreasing_intervals.append((-np.inf, roots[0]))
    else:
        decreasing_intervals.append((roots[0], np.inf))
else:
    decreasing_intervals.append((-np.inf, np.inf))
print("Интервалы убывания:", decreasing_intervals)
```

### 4. Построение графика
Стоим график функции `f(x)` на интервале от -10 до 10 с помощью функций `np.linspace()` и `plt.plot()`. Затем добавляются подписи осей и заголовок графика с помощью функций `plt.xlabel()`, `plt.ylabel()` и `plt.title()`. Наконец, с помощью `plt.grid(True)` добавляются сетка на графике, и с помощью `plt.show()` выводится график на экран.

```python
x = np.linspace(-10, 10, 100)
y = f(x)
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("График функции f(x) = 5x^2 + 10x - 30")
plt.grid(True)
plt.show()
```

### 5. Вычисление вершины
Вычисляется координата x вершины параболы с помощью формулы `-b / (2a)`, где **a** и **b** - коэффициенты при **x^2** и **x** соответственно. Затем вычисляется значение y в этой точке. Координаты вершины сохраняются в переменной `vertex`.

```python
vertex_x = -10 / (2 * 5)
vertex_y = f(vertex_x)
vertex = (vertex_x, vertex_y)
print("Вершина функции:", vertex)
```

### 6. Промежутки, на которых f > 0
Определяются промежутки, на которых значение функции больше нуля. Если имеется два корня, то функция больше нуля на интервале между корнями. Если имеется один корень, то функция больше нуля либо до корня (если корень отрицательный), либо после корня (если корень положительный). Если корней нет, то функция больше нуля на всей числовой прямой. Промежутки сохраняются в переменной `positive_intervals`.

```python
positive_intervals = []
if len(roots) == 2:
    positive_intervals.append((roots[0], roots[1]))
elif len(roots) == 1:
    if roots[0] < 0:
        positive_intervals.append((roots[0], np.inf))
    else:
        positive_intervals.append((-np.inf, roots[0]))
else:
    positive_intervals.append((-np.inf, np.inf))
print("Промежутки, на которых f > 0:", positive_intervals)
```

### 7. Промежутки, на которых f < 0
Определяются промежутки, на которых значение функции меньше нуля. Если имеется два корня, то функция меньше нуля до первого корня и после второго корня. Если имеется один корень, то функция меньше нуля либо до корня (если корень положительный), либо после корня (если корень отрицательный). Если корней нет, то функция меньше нуля на всей числовой прямой. Промежутки сохраняются в переменной `negative_intervals`.

```python
negative_intervals = []
if len(roots) == 2:
    negative_intervals.append((-np.inf, roots[0]))
    negative_intervals.append((roots[1], np.inf))
elif len(roots) == 1:
    if roots[0] < 0:
        negative_intervals.append((-np.inf, roots[0]))
    else:
        negative_intervals.append((roots[0], np.inf))
else:
    negative_intervals.append((-np.inf, np.inf))
print("Промежутки, на которых f < 0:", negative_intervals)
```

### 8. Построение графиков убывающей и возрастающей ветвей
1. Функция `subplots` из модуля `matplotlib.pyplot` создает фигуру с двумя сюжетами (графиками) в одной строке. `ax1` и `ax2` - это объекты-оси, на которых будут отображаться графики.
2. Метод `plot` рисует график функции для значений `x` и `y`. Для возрастающей ветви, используются только те значения `x` и `y`, для которых `x >= vertex_x` . Цвет графика задается параметром `'g'`, который означает зеленый. Методы `set_title` и `grid` устанавливают заголовок и сетку на графике `ax1`.
3. Метод `plot` рисует график функции для значений `x` и `y`. Для убывающей ветви, используются только значения `x` и `y`, для которых `x <= vertex_x` . Цвет графика задается параметром `'r'`, который означает красный. Методы `set_title` и `grid` устанавливают заголовок и сетку на графике `ax2`.

```python
# Создаем два сюжета рядом друг с другом
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# График убывающей ветви
ax1.plot(x[x <= vertex_x ], f(x[x <= vertex_x ]), 'r')
ax1.set_title("Убывающая ветвь f(x)")
ax1.grid()

# График возрастающей ветви
ax2.plot(x[x >= vertex_x ], f(x[x >= vertex_x ]), 'g')
ax2.set_title("Возрастающая ветвь f(x)")
ax2.grid()

plt.show()
```
